`8.`  How many lines, words, and characters are in the file `ram_tzu.txt`?

````python
def Count(file_name):
    """Counts the lines, words, and characters in file_name."""
    with open(file_name) as file_object:
        text = file_object.read()
    no_lines = len(text.split('\n'))
    no_words = len(text.split())
    no_chars = len(text)
    return no_lines, no_words, no_chars

def main(file_name="DataSets/ram_tzu.txt"):
    print("{} has {} lines, {} words, {} characters.".format(file_name, *Count(file_name)))
    
main()
          
            
````

`9.`  Make a new file, `ram_tzu_titled.txt` where the first character of each word is capitalized.  Check `?str.title` for help.

````python
def TitleFile(file_name):
    front, ext = file_name.rsplit('.', 1)
    write_file_name = front + "_titled." + ext
    with open(file_name) as read_file:
        with open(write_file_name, "w") as write_file:
            for read_line in read_file:
                write_file.write(read_line.title())
    return write_file_name
                
def PrintFile(file_name):
    print("{}:".format(file_name))
    with open(file_name) as read_file:
        for line in read_file:
            print(line, end='')
    print("----")
    
def main(file_name = "DataSets/ram_tzu.txt"):
    PrintFile(file_name)
    titled_file = TitleFile(file_name)
    PrintFile(titled_file)
    
main()
    
````

will give you a list of all the `ipynb` files in the current directory.

Make a generator, `GetFirstPounds()` that can be used like this:
```python
for notebook, first_pound in GetFirstPounds():
```
and use it to produce this output:

Note: Some of the Lab notebooks need to be read as unicode.


````python
import glob
def GetFirstPounds():
    for file_name in glob.glob("./*.ipynb"):
        with open(file_name, encoding="utf-8") as file_obj:
            for line in file_obj:
                if line.find("#") > -1:
                    yield file_name, line
                    break
                    
def ReportPounds():
    for notebook, first_pound in GetFirstPounds():
        print("{:>40}:{}".format(notebook, first_pound))
        
def main():
    ReportPounds()
    
main()
````

`11.` Make a function that reads a file and returns the largest number of the same character in a row, which characters have that many in a row.  If the text in the file is:  "You better look at that report as soon as possible.", the return is (2, "tos") because there are 2 in a row of t's, o's, and s's.

````python
def FindNumberOfRepeats(file_name):
    """Returns (most_in_a_row, which_chars)  where there are most_in_a_row
    of each character in which_chars in the file named in the call.
    """
    most_in_a_row = 0
    which_chars = ""
    most_now = 1
    old_char = None
    with open(file_name) as file_object:
        while True:
            char = file_object.read(1)
            if not char:
                break
            if not char.isalnum():
                old_char == ""
                most_now = 1
                continue
            if char == old_char:
                most_now += 1
            else:
                if most_now > most_in_a_row:
                    which_chars = old_char
                    most_in_a_row = most_now
                elif most_now == most_in_a_row != 0:
                    if old_char not in which_chars:
                        which_chars += old_char
                most_now = 1
            old_char = char
    return most_in_a_row, which_chars

def main(file_name="DataSets/small.txt"):
    most_in_a_row, which_chars = FindNumberOfRepeats(file_name)
    whiches = ["{}'s".format(ch) for ch in which_chars]
    if whiches:
        which_str = ', '.join(whiches[:-1]) + ", and {}".format(whiches[-1])
    else:
        which_str = "no characters"
    print("{} has {} in a row of {}.".format(file_name, most_in_a_row, which_str))

main()
main("DataSets/empty.txt")
````
