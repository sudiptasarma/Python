`1.` Use the dataset `"DataSets/currencies.csv"`.  The data are supposed to be currency exchange rates.  But how can there be negative exchange rates?  
>   a.  Get the data into a dataframe and get rid of the column with the most negative numbers.

```python
import pandas as pd
currencies = pd.read_csv("DataSets/currencies.csv")
currencies = currencies.set_index("Unnamed: 0")
currencies.index.name = "date"
currencies.tail()
```

```python
currencies[currencies < 0].info()
```

```python
currencies.drop("ddm", axis=1, inplace=True)
currencies
```

>  b.  The `-1.000`s are sentinals for no data.  Do what you need to do to make reasonable guesses at what the data should be.

```python
import numpy as np
currencies.replace(-1, np.nan, inplace=True)
```

```python
currencies.interpolate()
```

>  c.  Notice that the date strings are not daily.  Make them daily and give reasonable values for the data.

```python
type(currencies.index[0])
```

```python
new_index = [860100 + d for d in range(1, 32)]
new_index[:8]
```
