`1.` Here are some beverages:  "Water", "Tea", "Coffee".  And here are some more: "Milk", "Kefir", "Lemonade".

> a. Make two lists of 3 beverages each.

````python
some = ["Water", "Tea", "Coffee"]
more = ["Milk", "Kefir", "Lemonade"]
````

> b. Put your two lists into one big list in three different ways.

````python
all_drinks = some + more
all_drinks
````

````python
some += more
some
````

````python
some = ["Water", "Tea", "Coffee"]
some.extend(more)
some
````

> c. Add "Beer" to your big list in 3 different ways.

````python
all_drinks += "Beer"
all_drinks
````

````python
all_drinks = ['Water', 'Tea', 'Coffee', 'Milk', 'Kefir', 'Lemonade']
all_drinks += ["Beer"]
all_drinks
````

````python
all_drinks.extend(["Beer"])
all_drinks
````

````python
all_drinks.append("Beer")
all_drinks
````

````python
all_drinks += "Beer",
", ".join(all_drinks)
````

````python
all_drinks.insert(0, "Beer")
", ".join(all_drinks)
````
