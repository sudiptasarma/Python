`3.` How many times to you have to toss a coin to get 5 heads in a row?  

````python
import random
tosses = 0
head_count = 0
while head_count < 5:
    tosses += 1
    if random.choice(("heads", "tails")) == "heads":
        head_count += 1
    else:
        head_count = 0
print("It took {} coin tosses to get 5 heads in a row.".format(tosses))
    
````

`4.`  The grade scale for this class is in the chart below.  I don't expect less than B-. 

Translate that into code and use it to give a letter grade for 93%.

|Percent | Letter Grade|
|:-------|:------------|
| >= 93 | A |
| 90 - 92 | A-  |
| 88 - 89 | B+ |
| 83 - 87 | B |
| 80 - 82  | B- |


````python
percent = 94
if percent >= 93:
    print('A')
elif percent >= 90:
    print('A-')
elif percent >= 88:
    print('B+')
elif percent >= 83:
    print('B+')
elif percent >= 80:
    print('B-')
else:
    print("Don't let this happen.")
````
