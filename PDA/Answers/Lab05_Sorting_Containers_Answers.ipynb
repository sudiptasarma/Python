`2.`  Here is a list of names: ["Aaron", "Zoe", "Leonard", "Josephine", "Sebastian"]
    Produce this output:
        'Aaron, Josephine, Leonard, Sebastian, and Zoe.'    

````python
names = ["Aaron", "Zoe", "Leonard", "Josephine", "Sebastian"]
names.sort()
", ".join(name for name in names[:-1]) + ", and " + names[-1] + '.'
````
